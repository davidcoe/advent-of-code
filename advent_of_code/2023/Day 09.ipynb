```python
import collections
from itertools import islice


class OSIS:
    input_data:list[int]

    def __init__(self, input_data):
        self.input_data = [int(i) for i in input_data.split(' ')]
        self.next = self.calculate_differences(self.input_data)
        print(f'Next: {self.next}')

    @classmethod
    def calculate_differences(cls, input_data:list[int]):
        if not any(input_data):
            # Base case
            # print(f'Base case: {input_data}')
            return 0

        # https://docs.python.org/3/library/itertools.html#recipes
        # sliding_window
        n = 2
        it = iter(input_data)
        window = collections.deque(islice(it, n-1), maxlen=n)
        differences = []
        for x in it:
            window.append(x)
            differences.append(window[1] - window[0])

        # print(f'{input_data} => {differences}')

        calculate_differences = OSIS.calculate_differences(differences)
        # print(f'I should add {calculate_differences} to my last element {input_data[-1]}')
        return input_data[-1] + calculate_differences

    @classmethod
    def parse_input(cls, input_data):
        return [OSIS(data) for data in input_data.split('\n')]

print(f'Example answer: {sum([o.next for o in OSIS.parse_input(puzzle.examples[0].input_data)])}')
```
