```python
from aocd.models import Puzzle

puzzle = Puzzle(year=2024, day=7)

def test(method, input, expected):
    actual = method(input)
    if actual == expected:
        print(f'\t☑ - {method.__name__}({input}) = {expected} = {actual}')
    else:
        print(f'\t☐ - {method.__name__}({input}) = {expected} ≠ {actual}')
```

```python
def parse_input(input:str):
    return [(int(test_value), list(map(int, values.split()))) for line in input.split('\n') for test_value, values in [line.split(': ')]]

parse_input(puzzle.examples[0].input_data)
```

```python
import itertools
from functools import reduce

def calculate_total_calibration_result(input):

    total_calibration_result = 0

    for test_value, values in parse_input(input):

        all_operator_combinations = list(itertools.product(operators, repeat=len(values) - 1))

        for operator_combo in all_operator_combinations:
            operator = iter(operator_combo)
            
            result = reduce(lambda x, y: x + y if next(operator) == '+' else x * y, values)
            if result == test_value:
                total_calibration_result += test_value
                break

    return total_calibration_result
```

```python
import itertools
from functools import reduce

def calculate_total_calibration_result(input):
    operators = ['+', '*', '||']
    total_calibration_result = 0

    for test_value, values in parse_input(input):

        all_operator_combinations = list(itertools.product(operators, repeat=len(values) - 1))

        for operator_combo in all_operator_combinations:
            operator = iter(operator_combo)
            
            result = reduce(lambda x, y: x + y if (op := next(operator)) == '+' else x * y if op == '*' else int(str(x) + str(y)), values)
            if result == test_value:
                total_calibration_result += test_value
                break

    return total_calibration_result
```

```python
test(calculate_total_calibration_result, puzzle.examples[0].input_data, 11387)
```

```python
calculate_total_calibration_result(puzzle.input_data)
```
